libraries

```python
import tenseal as ts
import pickle
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

key generation

```python
context = ts.context(ts.SCHEME_TYPE.BFV, poly_modulus_degree=8192, plain_modulus=1032193, coeff_mod_bit_sizes=[60, 40, 40, 60])
context.global_scale = 2**40
context.generate_galois_keys()

```

model

```python
m="heart_model.pkl"
with open(m,'rb') as f:
        model_heart=pickle.load(f)
```

loading data

```python
X = pd.read_csv('x.csv')
Y = pd.read_csv('y.csv')
```

```python
X
```

```python
Y
```

```python
x_test=np.array(X)
Y_test=np.array(Y)
```

```python
x_test
```

encryption

```python
enc_x_test=[]
for sublist in x_test:
    enc_sub = ts.bfv_vector(context, sublist)
    enc_x_test.append(enc_sub)
enc_x_test = np.array(enc_x_test)
```

```python
enc_x_test
```

```python
Y_test
```

```python
Y_test = Y_test.flatten()
Y_test = Y_test.reshape(-1)
```

```python
Y_test
```

```python
coef = model_heart.coef_.flatten()
coef = coef.reshape(-1)
```

```python
coef
```

encrypting model coef

```python
enc_coef= ts.bfv_vector(context, coef)
```

```python
enc_int= ts.bfv_vector(context,model_heart.intercept_)
```

```python
enc_coef
```

```python
model_heart.intercept_
```

```python
enc_int
```

prediction

```python
prediction=[]

for x in enc_x_test:
    l=x.dot(enc_coef)
    prediction.append(l+enc_int)

```

```python
prediction
```

decryption

```python
decrypted_op=[]
for i in range (0,len(prediction)):
    if(prediction[i].decrypt()[0]>0):
        decrypted_op.append(1)
    else:
        decrypted_op.append(0)
```

```python
decrypted_op
```

```python
accuracy = accuracy_score(Y_test, decrypted_op)
```
