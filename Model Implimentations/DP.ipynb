```python
import numpy as np
import pickle
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python

def add_laplace_noise(data, epsilon, reduction_factor):
    sensitivity = 1  # Assuming sensitivity of the data
    original_scale = sensitivity / epsilon
    reduced_scale = original_scale * reduction_factor
    noisy_data = []
    for x in data:
        l=[]
        for element in x:
            l.append(element + np.random.laplace(scale=reduced_scale))
        noisy_data.append(l)
    return noisy_data
```

```python
m="heart_model.pkl"
with open(m,'rb') as f:
        model_heart=pickle.load(f)
```

```python
X = pd.read_csv('x.csv')
Y = pd.read_csv('y.csv')
```

```python
x_test=np.array(X)
Y_test=np.array(Y)
```

```python
x_test
```

```python
x_test=add_laplace_noise(x_test, 1, 0.5)
x_test=np.array(x_test)
```

```python
x_test
```

```python
coef = model_heart.coef_.flatten()
coef = coef.reshape(-1)
```

```python
prediction=[]

for x in x_test:
    l=model_heart.intercept_[0]
    l+=x.dot(coef)
    prediction.append(l)
```

```python
prediction
```

```python
op=[]
for i in range (0,len(prediction)):
    if(prediction[i]>0):
        op.append(1)
    else:
        op.append(0)
```

```python
op
```

```python
accuracy_score(Y_test,op)
```
