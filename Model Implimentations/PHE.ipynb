importing libraries

```python
import numpy as np
import pandas as pd
import pickle
import phe as paillier
from sklearn.metrics import accuracy_score
```

model 

```python
m="heart_model.pkl"
with open(m,'rb') as f:
        model_heart=pickle.load(f)
```

key generation

```python
public_key, private_key = paillier.generate_paillier_keypair()
```

loading data

```python
X = pd.read_csv('x.csv')
Y = pd.read_csv('y.csv')
```

```python
X
```

```python
Y
```

```python
x_test=np.array(X)
Y_test=np.array(Y)
```

```python
x_test
```

encryption of data

```python
enc_x_test=[]
for x in x_test:
    enc_sub=[]
    for i in x:
        enc_sub.append(public_key.encrypt(i))
    enc_x_test.append(enc_sub)
enc_x_test = np.array(enc_x_test)
```

```python
enc_x_test
```

```python
Y_test
```

```python
Y_test = Y_test.flatten()
Y_test = Y_test.reshape(-1)
```

```python
Y_test
```

```python
coef = model_heart.coef_.flatten()
coef = coef.reshape(-1)
```

```python
coef
```

```python
model_heart.intercept_[0]
```

prediction on encrypted data

```python
prediction=[]

for x in enc_x_test:
    l=model_heart.intercept_[0]
    for i in range(0,len(coef)):
        l+=(x[i]*coef[i])
    prediction.append(l)
```

```python
prediction
```

decryption

```python
decrypted_op=[]
for i in range (0,len(prediction)):
    if(private_key.decrypt(prediction[i])>0):
        decrypted_op.append(1)
    else:
        decrypted_op.append(0)
```

```python
decrypted_op
```

```python
accuracy = accuracy_score(Y_test, decrypted_op)
```
